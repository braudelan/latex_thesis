```python
import re

# script needs to except a tex file and capitalize it while ignoring comments and latex macros.
# after capitalizing the relevant sentences, text needs to be combined with rest of the tex (comments and macros).

def capitalize(text):
    '''capitalize every letter immediately after '.', '!' or '?'.

     does not capitalize first letter of a section or paragraph.
     '''`
    split_sentences = re.split('([.!?] *)', text)
    final = ''.join([i.capitalize() for i in split_sentences])
    print(final)

# def capitalize_tex_file(file_name):
    
#     with open(file_name, mode='r') as f:
        
```

```python
capitalized_lines = []
for line in f.readlines():
    capitalized_lines.append(capitalize(line))
    
    
```
